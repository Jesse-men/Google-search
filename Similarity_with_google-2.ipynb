```python
!pip install sentence-transformers
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd                        
from pytrends.request import TrendReq
import numpy as np
```

```python
pytrend = TrendReq()
df = pytrend.trending_searches(pn='hong_kong')
print(df)
df.head(20)
```

```python
google = np.array(df)
google = list(np.ravel(google))
google
```

```python
eyecream = ['撫紋',
 '早睡早起',
 '眼霜',
 '皺紋',
 '眼紋',
 '抗皺',
 '紋細紋',
 '修護',
 '光澤滋潤',
 '最強',
 '無紋',
 '護膚',
 '除眼袋',
 '滋潤眼',
 '睡覺']
```

```python
model = SentenceTransformer('distiluse-base-multilingual-cased-v1')
candidate_embeddings = model.encode(eyecream)
doc_embedding = model.encode(google)
distances = cosine_similarity(candidate_embeddings,doc_embedding)
```

```python
list1 = list(eyecream)
list2 = distances.tolist()
tuple1=zip(list1,list2)
List1=[]
for i in tuple1:
  List1.append(i)
print("keywords and similarities：", List1)
```
